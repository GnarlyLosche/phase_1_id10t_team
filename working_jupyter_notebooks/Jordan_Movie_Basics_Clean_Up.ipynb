```python
import csv
import json
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
conn = sqlite3.connect('../Data/im.db')
```

# Counting number of duplicate Primary_Titles in Movie Basics

```python
pd.read_sql('''
SELECT
    primary_title,
    COUNT(primary_title)
FROM
    movie_basics
GROUP BY
    primary_title
HAVING
    COUNT(primary_title) > 2

''',conn)
```

# SQL Subquery that Selects only Movies with a numvotes over 700

This is done under the assumption that the smaller movies that have the same name as larger movies will be removed.

```python
cleaned_movie_basics = pd.read_sql('''
SELECT
    *
FROM
    movie_basics
WHERE
    movie_id IN (SELECT movie_id
                 FROM movie_ratings
                 WHERE numvotes > 700)

''',conn)
```
